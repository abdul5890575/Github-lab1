Dropping not assigned values

```python
df=df[0]
```

```python
df=df[df.Borough != 'Not assigned']
df.reset_index(inplace=True)
df.drop('index',axis=1,inplace=True)
df.head(20)
```

#### Joining same postal code neighborhoods

```python
df.groupby(['Postcode','Borough'],sort=False).agg(lambda x: ','.join(x))
```

#### Replacing not assigned neighborhoods

```python
df['Neighborhood']=df['Neighborhood'].replace('Not assigned', df['Borough'])
```

```python
df2=pd.read_csv('http://cocl.us/Geospatial_data')                      
df2.head()
```

```python
res=df.merge(df2,left_on='Postcode',right_on='Postal Code')
res.drop('Postal Code',axis=1,inplace=True)
```
