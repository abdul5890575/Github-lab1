#### Importing libraries and bringing data

```python
import numpy as np 
import pandas as pd
from pandas.io.html import read_html
!pip install lxml
path="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df=pd.read_html(path, attrs={"class":"wikitable"})
```

Dropping not assigned values

```python
df=df[0]
```

```python
df=df[df.Borough != 'Not assigned']
df.reset_index(inplace=True)
df.drop('index',axis=1,inplace=True)
df.head(20)
```

#### Joining same postal code neighborhoods

```python
df.groupby(['Postcode','Borough'],sort=False).agg(lambda x: ','.join(x))
```

#### Replacing not assigned neighborhoods

```python
df['Neighborhood']=df['Neighborhood'].replace('Not assigned', df['Borough'])
```
